# question 1 - Explain the basic concept of clustering and give examples of applications where clustering is useful.

Clustering is a fundamental technique in unsupervised machine learning and data analysis that involves grouping similar data points together based on their shared characteristics or patterns. The core concept of clustering is to identify natural structures or groupings within a dataset without any prior knowledge of class labels. Here's a basic explanation of clustering and examples of applications where clustering is useful:

**Basic Concept of Clustering:**

- Clustering aims to partition a dataset into clusters, where data points within the same cluster are more similar to each other than to those in other clusters.

- The central idea is to maximize intra-cluster similarity (similarity among data points within the same cluster) and minimize inter-cluster similarity (similarity between data points from different clusters).

- Clustering algorithms utilize various similarity measures or distance metrics to determine how data points are grouped together.

**Examples of Applications Where Clustering Is Useful:**

1. **Customer Segmentation:**
   - **Application:** In marketing, businesses use clustering to segment their customer base into distinct groups based on purchasing behavior, demographics, or preferences. This helps tailor marketing strategies and product recommendations to specific customer segments.

2. **Image Segmentation:**
   - **Application:** In computer vision, clustering is employed for image segmentation, where similar pixels or regions in an image are grouped together. It's used in object recognition, medical imaging, and scene analysis.

3. **Anomaly Detection:**
   - **Application:** Clustering can be used for anomaly detection, identifying data points that deviate significantly from the norm. This is applied in fraud detection, network security, and quality control to spot unusual patterns or outliers.

4. **Document Clustering (Text Mining):**
   - **Application:** In natural language processing, clustering is used to group similar documents or articles together for tasks like topic modeling, document organization, and information retrieval.

5. **Recommendation Systems:**
   - **Application:** Clustering is integral to recommendation systems, where it helps group users or items with similar preferences. This aids in generating personalized recommendations for products, movies, or content.

6. **Genomic Data Analysis:**
   - **Application:** Clustering is used in genomics to analyze gene expression data. It helps discover patterns in gene activity, identify gene clusters with similar expression profiles, and uncover potential associations with diseases.

7. **Market Basket Analysis:**
   - **Application:** In retail, clustering is employed for market basket analysis, revealing associations between products frequently purchased together. Retailers use this information for stock placement, promotions, and product recommendations.

8. **Spatial Data Analysis:**
   - **Application:** In geographic information systems (GIS), clustering aids in identifying spatial patterns and hotspots in data. It's applied in epidemiology to locate disease clusters, urban planning to identify traffic congestion areas, and criminology to pinpoint crime hotspots.

9. **Biology and Taxonomy:**
   - **Application:** Clustering is used in biology to classify species into taxonomic groups based on shared characteristics or genetic traits. It helps create hierarchical classification systems.

10. **Image Compression:**
    - **Application:** Clustering can be employed in image compression to reduce the size of image files by grouping similar pixel values together, thus preserving image quality while reducing storage requirements.

These examples highlight the versatility of clustering in various domains, including marketing, computer vision, security, data analysis, and scientific research. Clustering provides valuable insights, aids in pattern recognition, and supports decision-making processes in a wide range of applications. The choice of clustering algorithm and similarity metric depends on the specific characteristics of the data and the objectives of the analysis.

# question 2 -- What is DBSCAN and how does it differ from other clustering algorithms such as k-means and hierarchical clustering?

DBSCAN, which stands for "Density-Based Spatial Clustering of Applications with Noise," is a clustering algorithm used in unsupervised machine learning to identify clusters in data based on the density of data points. DBSCAN is different from other clustering algorithms such as K-Means and Hierarchical Clustering in several ways:

**DBSCAN:**

1. **Density-Based Approach:**
   - DBSCAN groups data points based on their density. It defines clusters as areas of high data point density separated by areas of low density. Data points within a cluster are densely connected to each other, and data points in low-density regions are considered noise or outliers.

2. **Variable Cluster Shapes:**
   - DBSCAN can identify clusters with arbitrary shapes. It is not limited to finding spherical or convex clusters like K-Means. This flexibility makes it suitable for datasets with irregularly shaped clusters.

3. **No Need to Specify the Number of Clusters (K):**
   - In DBSCAN, you don't need to specify the number of clusters in advance, which is a requirement in K-Means. The algorithm automatically detects the number of clusters based on data density.

4. **Robust to Noise and Outliers:**
   - DBSCAN can effectively handle noisy data and outliers by classifying them as noise rather than assigning them to clusters. This robustness to noise is particularly valuable in real-world datasets.

5. **Hierarchical Clustering Structure:**
   - DBSCAN can uncover hierarchical clustering structures within the data. It identifies core points, border points, and noise points, creating a hierarchical relationship between them.

6. **Variable Cluster Density:**
   - DBSCAN can find clusters of varying densities, which is beneficial when dealing with datasets where some clusters are denser than others.

7. **Insensitive to Data Order:**
   - DBSCAN's results are insensitive to the order in which data points are processed, which is not the case for some other algorithms like K-Means.

**K-Means:**

1. **Centroid-Based Clustering:**
   - K-Means is a centroid-based clustering algorithm that partitions data into K clusters by minimizing the sum of squared distances from data points to cluster centroids. It assumes clusters with spherical shapes and similar sizes.

2. **Sensitive to Initialization:**
   - K-Means is sensitive to the initial placement of cluster centroids, which can lead to different solutions with different initializations.

3. **Assumes Equal Variance and Size:**
   - K-Means assumes that clusters have roughly equal variance and size, which may not hold for datasets with irregularly shaped or differently sized clusters.

4. **Euclidean Distance:**
   - K-Means typically uses the Euclidean distance metric to measure similarity between data points, which may not be suitable for all types of data.

**Hierarchical Clustering:**

1. **Hierarchy of Clusters:**
   - Hierarchical clustering creates a hierarchy of clusters, allowing you to explore different levels of granularity in clustering. It is not limited to a single partitioning of the data.

2. **Sensitive to Initialization:**
   - Like K-Means, agglomerative hierarchical clustering can be sensitive to the initial order in which data points are processed, which can affect the resulting hierarchy.

3. **Computationally Intensive:**
   - Hierarchical clustering can be computationally intensive, especially for large datasets, as it requires pairwise distance or similarity calculations for all data points.

In summary, DBSCAN is a density-based clustering algorithm that excels at finding clusters of varying shapes and sizes, automatically determining the number of clusters, and handling noisy data. K-Means is centroid-based and assumes spherical clusters of equal variance, while hierarchical clustering constructs a hierarchy of clusters and is sensitive to initialization. The choice of clustering algorithm should be based on the characteristics of the data and the specific objectives of the analysis.

# question 3 - How do you determine the optimal values for the epsilon and minimum points parameters in DBSCAN clustering?

Determining the optimal values for the epsilon (Îµ) and minimum points (MinPts) parameters in DBSCAN (Density-Based Spatial Clustering of Applications with Noise) can significantly impact the clustering results. The choice of these parameters depends on the characteristics of your data and the goals of your clustering analysis. Here's a general guideline on how to determine these values:

1. **Understand the Data:**
   - Begin by gaining a thorough understanding of your dataset. Consider its size, dimensionality, and the inherent distribution of data points. Visualize the data if possible.

2. **Select a Distance Metric:**
   - Choose an appropriate distance metric (e.g., Euclidean distance, Manhattan distance, or a custom metric) based on the nature of your data and the problem you are trying to solve.

3. **Set the Minimum Points (MinPts):**
   - The MinPts parameter determines the minimum number of data points required to form a dense region (core point). It depends on the expected density of clusters in your data. You can set MinPts based on domain knowledge or use a rule of thumb like:
     - MinPts = 2 * Dimensionality of the data (for low-dimensional data).
     - MinPts = 4 to 6 for higher-dimensional data.

4. **Select the Epsilon (Îµ) Value:**
   - The epsilon parameter defines the radius within which a data point must have at least MinPts neighboring points to be considered a core point. The choice of Îµ depends on the density of your data and the desired cluster sizes. Several methods can help you determine Îµ:

   - **Visual Inspection:** Plot a k-distance graph, where k is the number of nearest neighbors, and observe the point at which the graph starts to exhibit an "elbow" or a significant increase in distances. This point can be a good estimate of Îµ.

   - **Reachability Plot:** Compute the reachability distance for each point in the dataset based on MinPts and plot these distances. Look for a point where the reachability distances suddenly increase, indicating the start of a new cluster.

   - **Silhouette Score:** Experiment with different Îµ values and compute the silhouette score for the resulting clusters. Choose the Îµ that yields the highest silhouette score, indicating better cluster quality.

   - **Grid Search:** Perform a grid search over a range of Îµ values and evaluate the clustering results using internal validation metrics or domain-specific criteria.

   - **Domain Knowledge:** In some cases, domain knowledge or prior information about the expected cluster sizes or distances between data points can guide the choice of Îµ.

5. **Evaluate and Refine:**
   - After setting Îµ and MinPts, run DBSCAN and evaluate the clustering results. Use internal validation metrics (e.g., silhouette score, Davies-Bouldin index) and visual inspection to assess the quality of the clusters. If necessary, refine the parameter values and repeat the process until you are satisfied with the results.

6. **Consider Robustness:** Keep in mind that DBSCAN is generally robust to small variations in parameter values. Therefore, it's essential to choose parameters that produce meaningful and interpretable clusters while considering the specific characteristics of your data.

7. **Interpret the Results:** Carefully interpret the clusters obtained with the chosen parameters and ensure they align with the objectives of your analysis.

The process of determining optimal Îµ and MinPts values may require some experimentation and domain knowledge, as there is no one-size-fits-all solution. It's essential to strike a balance between detecting meaningful clusters and avoiding over-segmentation or under-segmentation of your data.

# question 4 - How does DBSCAN clustering handle outliers in a dataset?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) clustering handles outliers in a dataset by classifying them as noise points that do not belong to any cluster. This capability makes DBSCAN particularly robust in the presence of noisy data. Here's how DBSCAN identifies and handles outliers:

1. **Core Points:**
   - DBSCAN categorizes data points into three types: core points, border points, and noise points. Core points are data points located within dense regions, meaning they have at least "MinPts" neighboring data points (including themselves) within a radius of "Îµ."

2. **Border Points:**
   - Border points are data points that have fewer than "MinPts" neighboring points within Îµ but are reachable from a core point. In other words, they are on the fringes of clusters but are not considered core points themselves.

3. **Noise Points (Outliers):**
   - Noise points, also known as outliers, are data points that do not meet the criteria to be classified as core points or border points. Specifically, they have fewer than "MinPts" neighboring points within Îµ and are not reachable from any core point.

How DBSCAN handles outliers:

- Noise points are not assigned to any cluster. Instead, they are treated as standalone data points that do not belong to any group or cluster.

- By design, DBSCAN does not force every data point to be assigned to a cluster, making it robust to the presence of noisy or anomalous data points. Noise points are allowed to exist as individual entities in the output.

- Outliers are identified as data points that are distant from dense clusters or do not conform to the density-based clustering pattern. They are typically isolated data points, distant from any cluster core.

- The noise points can be accessed separately in the clustering output, allowing users to identify and potentially investigate these outliers further, as they may represent interesting or unusual data instances.

In summary, DBSCAN handles outliers by classifying them as noise points that do not belong to any cluster. This approach makes DBSCAN a suitable choice for clustering tasks where the dataset may contain noisy or irregularly distributed data points, as it can effectively identify and segregate them from meaningful clusters.

# question 5 - How does DBSCAN clustering differ from k-means clustering?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) clustering and K-Means clustering are two fundamentally different clustering algorithms with distinct approaches and characteristics. Here are the key differences between DBSCAN and K-Means clustering:

1. **Clustering Approach:**

   - **DBSCAN:** DBSCAN is a density-based clustering algorithm. It groups data points based on their density, aiming to find regions of high data point density separated by regions of lower density. It defines clusters as dense areas in the data space.

   - **K-Means:** K-Means is a centroid-based clustering algorithm. It partitions the data into K clusters by minimizing the sum of squared distances between data points and cluster centroids. It aims to create clusters around central points, making it suitable for spherical or convex clusters.

2. **Number of Clusters:**

   - **DBSCAN:** DBSCAN does not require you to specify the number of clusters (K) in advance. It automatically detects the number of clusters based on the data's density and does not enforce that all data points belong to clusters.

   - **K-Means:** K-Means requires you to specify the number of clusters (K) before running the algorithm. This can be a limitation when you don't know the ideal number of clusters in your data.

3. **Cluster Shape:**

   - **DBSCAN:** DBSCAN can identify clusters of arbitrary shapes, including irregular, non-convex clusters. It is not constrained by assumptions about cluster shapes, making it suitable for various data distributions.

   - **K-Means:** K-Means assumes that clusters have a spherical or convex shape and similar sizes. It may struggle to identify clusters with complex or irregular shapes.

4. **Handling Outliers:**

   - **DBSCAN:** DBSCAN is robust to outliers. It classifies outliers as noise points that do not belong to any cluster. Outliers do not affect the formation of clusters in DBSCAN.

   - **K-Means:** K-Means is sensitive to outliers. Outliers can significantly impact the position and quality of cluster centroids, potentially leading to suboptimal clustering results.

5. **Parameter Dependency:**

   - **DBSCAN:** DBSCAN depends on two main parameters, Îµ (epsilon) and MinPts, which determine the density and neighborhood size used for clustering. While these parameters can impact results, DBSCAN is relatively robust to their variations.

   - **K-Means:** K-Means depends on the initial placement of cluster centroids, which can affect the final clustering results. It is sensitive to initialization and may converge to different solutions with different initializations.

6. **Handling Unevenly Sized Clusters:**

   - **DBSCAN:** DBSCAN can handle clusters of varying sizes naturally. Clusters can have different densities, and DBSCAN can adapt to these variations.

   - **K-Means:** K-Means tends to create clusters of roughly equal sizes and variances. It may struggle with datasets containing clusters of vastly different sizes.

7. **Scalability:**

   - **DBSCAN:** DBSCAN can be computationally more intensive, especially for large datasets, as it involves distance calculations between data points. It may not be as efficient as K-Means for very high-dimensional data.

   - **K-Means:** K-Means can be more scalable and computationally efficient, making it suitable for large datasets, especially when using optimizations like mini-batch K-Means.

In summary, DBSCAN and K-Means are different clustering algorithms that excel in different scenarios. DBSCAN is ideal for identifying clusters of arbitrary shapes and handling noisy data, while K-Means is suitable for finding spherical or convex clusters when the number of clusters is known in advance. The choice between these algorithms depends on the data characteristics and the specific objectives of the clustering task.

# question 6 - Can DBSCAN clustering be applied to datasets with high dimensional feature spaces? If so, what are some potential challenges?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) can be applied to datasets with high-dimensional feature spaces, but it may face several challenges in such scenarios. While DBSCAN is a versatile clustering algorithm, its effectiveness in high-dimensional spaces depends on various factors. Here are some considerations and challenges when applying DBSCAN to high-dimensional data:

1. **Curse of Dimensionality:** High-dimensional spaces are susceptible to the "curse of dimensionality." As the number of dimensions increases, the distance between data points becomes less discriminative, and data points appear more uniformly distributed in space. This can make it challenging to define meaningful density thresholds for DBSCAN.

2. **Increased Computational Complexity:** Calculating distances between data points in high-dimensional spaces is computationally more expensive. The time complexity of DBSCAN is typically O(n^2) for dense datasets, where n is the number of data points. In high-dimensional spaces, this computational cost can become a limiting factor, making DBSCAN less efficient.

3. **Parameter Sensitivity:** DBSCAN relies on the Îµ (epsilon) parameter, which defines the radius for neighborhood searches. In high-dimensional spaces, determining an appropriate Îµ value can be challenging because the distances between data points tend to be similar, making it difficult to choose a suitable density threshold.

4. **Sparse Data:** High-dimensional datasets are often sparse, meaning that most feature values are zero or near-zero. This sparsity can lead to difficulties in defining neighborhood relationships, as data points may appear distant due to the presence of many zero-valued features.

5. **Dimension Reduction:** Dimension reduction techniques, such as Principal Component Analysis (PCA) or t-Distributed Stochastic Neighbor Embedding (t-SNE), may be necessary to reduce the dimensionality of the data before applying DBSCAN. This can help mitigate the curse of dimensionality and improve clustering performance.

6. **Noise Sensitivity:** In high-dimensional spaces, the presence of noise and outliers can have a more significant impact on clustering results. DBSCAN's ability to handle noise remains an advantage, but defining what constitutes noise can be challenging.

7. **Interpretability:** Clustering results in high-dimensional spaces may be difficult to interpret and visualize, especially when dealing with many dimensions. Visualizing clusters or understanding their characteristics becomes more complex.

8. **Feature Scaling:** It's crucial to scale or normalize features in high-dimensional spaces to ensure that all dimensions contribute equally to the distance calculations. Failure to do so may result in certain dimensions dominating the clustering process.

9. **Domain Knowledge:** Incorporating domain knowledge or insights into the clustering process can be essential for selecting appropriate parameters and making sense of the results, especially in high-dimensional settings.

In summary, while DBSCAN can be applied to high-dimensional datasets, it faces challenges related to the curse of dimensionality, computational complexity, parameter sensitivity, and interpretability. Careful preprocessing, feature engineering, and parameter tuning, along with considerations for the specific characteristics of the data, are essential when using DBSCAN in high-dimensional feature spaces. In some cases, dimension reduction or alternative clustering algorithms designed for high-dimensional data may be more suitable.

# question 7 - How does DBSCAN clustering handle clusters with varying densities?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is particularly well-suited for handling clusters with varying densities. Unlike some other clustering algorithms that assume clusters have similar densities, DBSCAN can identify clusters of different densities effectively. Here's how DBSCAN handles clusters with varying densities:

1. **Core Points:** DBSCAN categorizes data points into three types: core points, border points, and noise points. Core points are data points that have at least "MinPts" neighboring data points (including themselves) within a radius of "Îµ."

2. **Border Points:** Border points are data points that have fewer than "MinPts" neighboring points within Îµ but are reachable from a core point. In other words, they are on the fringes of clusters but are not considered core points themselves.

How DBSCAN handles clusters with varying densities:

- DBSCAN defines clusters based on the concept of "density-reachability." A data point is considered part of a cluster if it can be reached from another data point by following a path of "density-reachable" core points.

- Dense regions with a high concentration of core points form the core of a cluster. These dense regions can have various shapes and sizes, and DBSCAN effectively identifies them.

- Border points on the fringes of clusters may belong to clusters with lower densities. They are linked to core points but do not contribute to the core of the cluster.

- Noise points, which do not satisfy the density criteria for any cluster, are treated as outliers and are not assigned to any cluster. This allows DBSCAN to effectively handle regions of low data point density.

- As a result, DBSCAN can uncover clusters with varying densities in the data without assuming that all clusters have uniform density. It adapts to the local density characteristics of the data.

- The choice of the Îµ (epsilon) and MinPts parameters in DBSCAN is crucial for detecting clusters with varying densities. Smaller Îµ values will yield denser clusters, while larger Îµ values will allow for the identification of clusters with lower densities.

In summary, DBSCAN's ability to identify core points, border points, and noise points based on data density allows it to naturally handle clusters with varying densities. This flexibility makes DBSCAN well-suited for datasets containing clusters of different sizes and shapes, where other clustering algorithms that assume uniform cluster densities may struggle.

# question 8 - What are some common evaluation metrics used to assess the quality of DBSCAN clustering results?

Evaluating the quality of DBSCAN (Density-Based Spatial Clustering of Applications with Noise) clustering results is essential to assess how well the algorithm has performed on your data. Several evaluation metrics can be used to measure the quality of DBSCAN clustering results. The choice of a specific metric depends on the characteristics of your data and the goals of your analysis. Here are some common evaluation metrics for DBSCAN:

1. **Silhouette Score:**
   - The Silhouette Score measures how similar each data point in one cluster is to the other data points in the same cluster compared to the nearest neighboring cluster. It ranges from -1 to 1, with higher values indicating better clustering. A higher Silhouette Score suggests that the data points are well-separated into clusters.

2. **Davies-Bouldin Index:**
   - The Davies-Bouldin Index evaluates the average similarity between each cluster and its most similar cluster. Lower values of the index indicate better clustering, with a minimum value of 0 indicating perfect clustering.

3. **Dunn Index:**
   - The Dunn Index assesses the ratio of the minimum inter-cluster distance to the maximum intra-cluster distance. A higher Dunn Index suggests better clustering, with larger inter-cluster gaps and smaller intra-cluster distances.

4. **Adjusted Rand Index (ARI):**
   - The Adjusted Rand Index measures the similarity between the true class labels (if available) and the cluster assignments. It adjusts for chance and yields a score between -1 and 1, with higher values indicating better agreement between the clustering and the true labels.

5. **Normalized Mutual Information (NMI):**
   - The Normalized Mutual Information measures the mutual information between the true labels and the predicted clusters, normalized by entropy. It provides a score between 0 and 1, with higher values indicating better agreement between clustering and true labels.

6. **Calinski-Harabasz Index (Variance Ratio Criterion):**
   - The Calinski-Harabasz Index is a ratio of between-cluster variance to within-cluster variance. Higher values suggest better clustering, indicating well-separated clusters with tight data point groupings.

7. **Within-Cluster Sum of Squares (WCSS):**
   - Although commonly used for K-Means, the WCSS can also be used to evaluate the quality of DBSCAN clustering. It measures the sum of squared distances between data points and their cluster centroids. Lower values indicate better clustering.

8. **Contingency Table Metrics (e.g., Completeness and Homogeneity):**
   - Completeness and homogeneity are two metrics that measure different aspects of cluster quality. Completeness assesses whether all data points belonging to the same true class are assigned to the same cluster, while homogeneity measures whether all data points within a cluster belong to the same true class.

9. **Visual Inspection:**
   - Sometimes, the best way to evaluate DBSCAN clustering results is through visual inspection. Plotting the clusters and examining how well they separate data points can provide valuable insights into the quality of the clustering.

10. **Domain-Specific Metrics:**
    - Depending on the specific problem and domain, you may need to define custom evaluation metrics that align with your objectives and the characteristics of your data.

It's important to note that DBSCAN may not always perform well with traditional clustering evaluation metrics, especially if your data contains irregularly shaped clusters or varying densities. Therefore, a combination of multiple metrics, visual inspection, and domain knowledge is often necessary to comprehensively assess the quality of DBSCAN clustering results.

   - Once you have the initial clusters from DBSCAN, you can use the labeled data (if available) to propagate labels to the data points within each cluster. This can be done in various ways, such as assigning the majority class label within each cluster to all data points in that cluster.

4. **Training and Evaluation:**

It's essential to keep in mind the following considerations when using DBSCAN in a semi-supervised learning context:

- The effectiveness of this approach depends on the quality and quantity of labeled data. If you have a small amount of labeled data, the propagation of labels may have limited impact. Additionally, noisy or mislabeled data points can affect the quality of the propagated labels.

- Feature selection and engineering are critical steps in improving the performance of the semi-supervised learning model. Carefully consider which features to use, as well as how to represent the propagated labels.

- Experiment with different strategies for label propagation, as well as various semi-supervised learning algorithms, to find the most effective combination for your specific task.

In summary, while DBSCAN is primarily an unsupervised clustering algorithm, it can be used as a preprocessing step in a semi-supervised learning pipeline. By propagating labels from clusters to data points and incorporating this information into a semi-supervised model, you can leverage the benefits of both clustering and supervised learning for semi-supervised tasks.

5. **Scalability:** Handling missing values can increase the computational complexity of DBSCAN, especially if imputation involves calculating distances between data points with missing values. Be aware of potential scalability issues.

6. **Robustness:** Evaluate the robustness of your clustering results to missing values. Assess how different imputation strategies and handling of missing values affect the clustering outcome.

In summary, DBSCAN is capable of handling datasets with noise, and it explicitly classifies noisy data points as outliers. When dealing with missing values, preprocessing steps like imputation, assigning special values, and careful selection of distance metrics can be employed to make the data suitable for DBSCAN. However, it's essential to consider the impact of these preprocessing steps on the clustering results and to experiment with different strategies based on the specific characteristics of your dataset.

# question 11 - Implement the DBSCAN algorithm using a python programming language, and apply it to a sample dataset. Discuss the clustering results and interpret the meaning of the obtained clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generate a sample dataset with two moons
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X)

# Visualize the clustering results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("DBSCAN Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Interpretation of Clusters
unique_labels = np.unique(labels)
n_clusters = len(unique_labels)  
print(f"Number of clusters: {n_clusters}")

for cluster_label in unique_labels:
    cluster_points = X[labels == cluster_label]
    cluster_size = len(cluster_points)
    cluster_center = np.mean(cluster_points, axis=0)
    print(f"Cluster {cluster_label}: Size={cluster_size}, Center={cluster_center}")
```
